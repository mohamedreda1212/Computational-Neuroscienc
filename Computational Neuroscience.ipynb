```python
from tkinter import Tk, Label, Entry, Button


def random():
    return (hash(str(id(object()))) % 1000) / 1000 - 0.5

def tanh(x):
    if x < -1:
        return -1
    elif x > 1:
        return 1
    else:
        return x  


def compute_output():
    try:
       
        x1 = float(entry_x1.get())
        x2 = float(entry_x2.get())

        
        w1_1 = random()
        w1_2 = random()
        w2_1 = random()
        w2_2 = random()
        w3_1 = random()
        w3_2 = random()

        
        b1 = 0.5
        b2 = 0.7

       
        h1 = tanh(w1_1 * x1 + w2_1 * x2 + b1)
        h2 = tanh(w1_2 * x1 + w2_2 * x2 + b1)

       
        y = tanh(w3_1 * h1 + w3_2 * h2 + b2)

        
        label_output.config(text=f"Output: {y:.4f}", fg="black")
    except ValueError:
        label_output.config(text="Error: Enter valid numbers!", fg="red")


root = Tk()
root.title("Neural Network GUI")
root.geometry("400x400")
root.configure(bg="#f0f0f0")


Label(root, text="Input x1:", bg="#f0f0f0").pack()
entry_x1 = Entry(root)
entry_x1.pack()

Label(root, text="Input x2:", bg="#f0f0f0").pack()
entry_x2 = Entry(root)
entry_x2.pack()


Button(root, text="Compute Output", command=compute_output, bg="#4CAF50", fg="white").pack(pady=10)


label_output = Label(root, text="Output: ", bg="#f0f0f0", font=("Arial", 12))
label_output.pack()


root.mainloop()
```
